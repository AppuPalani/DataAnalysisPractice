```python
x = ["a", "b", "c", "d" ,"e"]
```

```python
y = [45,48,49,78,25]
```

```python
z = [58,78,59,89,69]
```

```python
plt.plot(x,y,"o",c="y",ms=9,ls="--")
plt.grid()
plt.title("God is Great", color= "green", fontsize = 15)
plt.xlabel("No of students")
plt.ylabel("Marks of students")
plt.plot(x,z,"gs:")
plt.legend(["Marks in Maths", "Marks in Python"])
```

```python
plt.bar(x,y, color="skyblue", alpha = 0.8)
```

```python
plt.barh(x,y, color="skyblue", alpha = 0.5)
```
