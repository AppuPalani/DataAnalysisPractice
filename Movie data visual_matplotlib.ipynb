```python
#pip install matplotlib
import matplotlib.pyplot as plt
import pandas as pd
# Load the data
data = pd.read_csv('Movies.csv')

movie_count = data['movie_title'].value_counts().sum()
print(movie_count)

#To plot the distribution of movies releases over the years
plt.figure(figsize=(6,4))
plt.hist(data['title_year'], bins=30, edgecolor='k')
plt.xlabel('Release Year')
plt.ylabel('Number of Movies')
plt.title('Distribution of movies released over the years')
plt.grid(True)
plt.show()
```

```python
# Find the top 10 movie genres and their frequencies

# count the movie genres and get top 10
genre_counts = data['genre_1'].value_counts().head(10)

#plotting the top 10 movie genres

plt.figure(figsize=(6,4))
genre_counts.plot(kind='bar')
plt.xlabel('Genre')
plt.ylabel('Number of Movies')
plt.title('Top 10 movie genres')
plt.grid(True)
plt.show()
```

```python
# Distribution of movie origins
# count the movie origins and get top 5
origin_counts = data['country'].value_counts().head(5)

#plotting the top 10 movie genres
plt.figure(figsize=(6,4))
origin_counts.plot(kind='pie', autopct='%1.1f%%')
plt.title('Top 5 movie origin country')
plt.show()
```
